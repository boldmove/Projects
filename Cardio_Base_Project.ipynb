```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded['cardio_base (1).csv']))
```

```python
df.head(10)
```

```python
# prompt: convert the values of the age column into ages in years rounded down

df['age'] = (df['age'] / 365.25).astype(int)
df.head(10)
```

```python
# prompt: in percentage, how much heavier is the age group with the highest average weight heavier than the age group with the lowest weight

avg_weight_by_age = df.groupby('age')['weight'].mean()
highest_avg_weight = avg_weight_by_age.max()
lowest_avg_weight = avg_weight_by_age.min()
percentage_difference = ((highest_avg_weight - lowest_avg_weight) / lowest_avg_weight) * 100
print(f'{percentage_difference:.2f}%')
```

```python
# prompt: Do people over 50 have higher cholestrol than the rest, what percentage higher or lower

over_50_chol = df[df['age'] > 50]['cholesterol'].mean()
under_50_chol = df[df['age'] <= 50]['cholesterol'].mean()

percentage_difference = ((over_50_chol - under_50_chol) / under_50_chol) * 100

print(f'{percentage_difference:.2f}%')
```

```python
# prompt: how many times are men likely to be smokers than women or viceversa

smoker_men = df[(df['gender'] == 2) & (df['smoke'] == 1)].shape[0]
smoker_women = df[(df['gender'] == 1) & (df['smoke'] == 1)].shape[0]

likelihood = smoker_men / smoker_women

print(f'{likelihood:.2f}')
```

```python
# prompt: how tall are the tallest 1%

tallest_one_percent = df['height'].quantile(0.99)
print(f'{tallest_one_percent:.2f}')
```

```python
# prompt: which two features have the highest spearman rank correlation. return the feature names

df.corr(method='spearman').unstack().sort_values(ascending=False).drop_duplicates().head(2)
```

```python
# prompt: % of those 2 SD from average height

from scipy import stats

std_height = df['height'].std()
mean_height = df['height'].mean()

# Calculate the bounds for 2 standard deviations from the mean
upper_bound = mean_height + 2 * std_height
lower_bound = mean_height - 2 * std_height

# Find the number of people outside these bounds
outside_2sd = df[(df['height'] > upper_bound) | (df['height'] < lower_bound)].shape[0]

# Calculate the percentage
percentage = (outside_2sd / df.shape[0]) * 100

print(f'{percentage:.2f}%')
```

```python
uploaded = files.upload()

df2 = pd.read_csv(io.BytesIO(uploaded['cardio_alco.csv']))
```

```python
df2.head(10)
```

```python
# prompt: use seperate columns i.e seperate id from alco in id;alco for df2

df2[['id', 'alco']] = df2['id;alco'].str.split(';', expand=True)
df2.head(10)
```

```python
# prompt: merge the two dataframe while ignoring ID with no alchol consuption information

# Convert 'id' column to numeric in both DataFrames
df['id'] = pd.to_numeric(df['id'])
df2['id'] = pd.to_numeric(df2['id'])

# Merge the DataFrames based on 'id', keeping only matching IDs
merged_df = pd.merge(df, df2, on='id', how='inner')

merged_df.head(10)
```

```python
# prompt: what % of people over 50 year old consume alcohol

# Calculate the percentage of people over 50 who consume alcohol
over_50_alcohol_percentage = (merged_df[(merged_df['age'] > 50) & (merged_df['alco'] == '1')].shape[0] / merged_df[merged_df['age'] > 50].shape[0]) * 100

# Print the result
print(f'{over_50_alcohol_percentage:.2f}%')
```

```python
# prompt: confidence rate of true that nonsmokers weigh less than smokers

# Separate weights of smokers and non-smokers
smoker_weights = merged_df[merged_df['smoke'] == 1]['weight']
nonsmoker_weights = merged_df[merged_df['smoke'] == 0]['weight']

# Perform t-test
t_stat, p_value = stats.ttest_ind(nonsmoker_weights, smoker_weights)

# Calculate confidence interval
confidence_level = 0.95
degrees_of_freedom = len(smoker_weights) + len(nonsmoker_weights) - 2
confidence_interval = stats.t.interval(confidence_level, degrees_of_freedom, loc=np.mean(nonsmoker_weights) - np.mean(smoker_weights), scale=stats.sem(nonsmoker_weights) + stats.sem(smoker_weights))

# Print results
print(f'T-statistic: {t_stat:.2f}')
print(f'P-value: {p_value:.3f}')
print(f'{confidence_level*100:.0f}% Confidence interval: {confidence_interval}')
```
